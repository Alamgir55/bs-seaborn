```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df
```

```python
sns.scatterplot(x='salary',y='sales',data=df)
plt.show()
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df)
plt.show()
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df, hue='level of education')
plt.show()
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df, hue='salary')
plt.show()
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df, hue='level of education', palette='Dark2')
plt.show()
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df, size='work experience')
plt.show()
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df, size=200,alpha=0.2)
plt.show()
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df, style='level of education', hue='level of education')
plt.show()
plt.savefig('my_plot.jpg')
```
